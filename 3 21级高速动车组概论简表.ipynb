```python
a2 = pd.read_excel(os.path.join(path,'2021 - 高速动车组概论-wzj.xlsx'),header=None)
a2
```

```python
header = ['班级','学号','姓名','性别','课程名称','课程编号','考试/考查',
          '1.1','1.2','1.3','1.4','2.1','2.2','2.3','3.1','3.2','4.1','4.2','4.3','5.1','5.2',
         '6.1','6.2','7.1','7.2','8.1','8.2','9.1','9.2','10.1','10.2','11.1','11.2','12.1','12.2',
         '总目标','目标1','目标2','目标3']
len(header)
```

```python
a22 = a2.iloc[2:,:39]
#增加列名：
a22.columns = header
a22
```

```python
#使a22行名从0开始：
a22.index = range(a22.shape[0])
a22
```

```python
#通过列名核对，将一个excel表的内容填到另一个excel表里面去。
name1 = a1.iloc[:,0]
name2 = a22.iloc[:,2] #找出姓名所在列
for i in range(len(name1)):
    for j in range(len(name2)):
        if name1[i] == name2[j]:
            a22.iloc[j,14] = a1.iloc[i,5] #3.1
            a22.iloc[j,21] = a1.iloc[i,6] #6.1
            a22.iloc[j,22] = a1.iloc[i,7] #6.2
            a22.iloc[j,35] = a1.iloc[i,4] #总目标
            a22.iloc[j,36] = a1.iloc[i,1] #目标1
            a22.iloc[j,37] = a1.iloc[i,2] #目标2
            a22.iloc[j,38] = a1.iloc[i,3] #目标3
```

```python
excel_file = os.path.join(path,'output6.xlsx')
with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
    a22.to_excel(writer, index=True, sheet_name='Sheet1')
```
